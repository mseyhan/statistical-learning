```python
import datetime
import numpy as np
import pandas as pd

report_name = "C69"
start_date= "01.01.2023" #"%d.%m.%Y"
end_date= "31.12.2023" #"%d.%m.%Y"
period = "D"
day_of = 13
holidays = pd.read_excel('holidays.xlsx', index_col=None, header=0)
holidays['CALENDAR_DATE']= pd.to_datetime(holidays['CALENDAR_DATE'],format="%d/%m/%Y").astype('datetime64[ns]')
#print(holidays)
holidays= holidays['CALENDAR_DATE'].tolist()
start_date = datetime.datetime.strptime(start_date, "%d.%m.%Y")
end_date = datetime.datetime.strptime(end_date, "%d.%m.%Y")
start_year = start_date.year
holidays
```

```python
def send_schedule_datemaker():

    if period == "M":
        
        
        
        #dq_send_date col
        dq_send_date = pd.bdate_range(start=start_date,end=end_date,freq=period,name="DQ_SEND_DATE").to_series()
        dq_send_date = np.array(dq_send_date.dt.to_period(period).astype(str))
        dq_send_date = pd.Series(dq_send_date)
        dq_send_date = dq_send_date.apply(lambda x: np.busday_offset(x, day_of-1, roll='forward',holidays=holidays))
        dq_send_date = dq_send_date.to_frame(name="DQ_SEND_DATE")
        #misc
        
        
    elif period == "D":
        
        
        
        #dq_send_date col
        dq_send_date = pd.bdate_range(start=start_date,end=end_date,freq="C",name="DQ_SEND_DATE"
                                      ,weekmask=[1,1,1,1,1,0,0],holidays=holidays)
        dq_send_date = pd.DataFrame(dq_send_date)
    
    cols = []
    df = pd.DataFrame(dq_send_date)

    # Create a Pandas Excel writer using XlsxWriter as the engine.
    writer = pd.ExcelWriter('C:/Users/MuratSey/Desktop/DQ/send_schedule/{report_name}_{suffix}.xlsx'
                            .format(report_name=report_name,suffix=start_year), engine='xlsxwriter')

    # Convert the dataframe to an XlsxWriter Excel object.
    df.to_excel(writer, sheet_name='Sheet1')

    # Close the Pandas Excel writer and output the Excel file.
    writer.save()
    return dq_send_date

send_schedule_datemaker()
```
